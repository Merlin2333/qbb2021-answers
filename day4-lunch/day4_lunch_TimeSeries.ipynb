## Load packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Process data

```python
df_fpkms = pd.read_csv( "../../qbb2021/data/fpkms.csv", index_col = "t_name") # explicitly identify the col name

df_replic = pd.read_csv( "../../qbb2021/data/replicates.csv", index_col = "t_name") # load the replicate file

df_replic
```

```python
df_sxl_male = df_fpkms.loc["FBtr0331261", "male_10":"male_14D"]# Get FBtr0331261 male expression data
df_sxl_female = df_fpkms.loc["FBtr0331261", "female_10":"female_14D"]# Get FBtr0331261 female expression data

df_replic_male = df_replic.loc["FBtr0331261", "male_14A":"male_14D"]
df_replic_female = df_replic.loc["FBtr0331261", "female_14A":"female_14D"]
```

```python
x = ['10', '11', '12', '13', '14A', '14B', '14C', '14D']
x_replic = ['14A', '14B', '14C', '14D']
```

## Plot

```python
fig, ax = plt.subplots()
ax.plot(x, df_sxl_male, label = "Male")

ax.scatter(x_replic, df_replic_male, label = "Male replicates", marker = '+', s = 100, c = 'red') 

ax.plot(x,df_sxl_female, label = "Female")

ax.scatter(x_replic, df_replic_female, label = "Female replicates", marker = '+', s = 100, c = 'purple')

ax.set_title ("Sex difference in the expression of FBtr0331261 in different stages", loc = 'center')
ax.set_xlabel("Developmental stage")
ax.set_ylabel("mRNA abundance (RPKM)")
ax.legend()
fig.savefig("../../qbb2021-answers/day4-lunch/Time_series.png")
plt.show()
```
