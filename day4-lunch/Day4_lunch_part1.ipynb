# Introduction

This exercise aims to create an MA plot to visualize RNA-seq data with the average gene abundance on the x-axis and fold change on the y-axis. An MA plot visualizes the differences between measurements taken in two samples. "A" is the average average intensity of each gene expression, as calculated by $$A = \frac{1}{2}(log_2(sample1) + log_2(sample2))$$ "M" is the log ratio of the gene expression intensity that indicates the difference expression levels between two genes. "M" is calculated as $$A = log_2(sample1) - log_2(sample2)$$ 

The equations and above overview is referenced at [MA plot](https://en.wikipedia.org/wiki/MA_plot).

This exercise compares the transcripts expressions between male_10 and male_14D. It is interesting to see possible differences in gene expression levels in early and late stages in male D. melanogaster embryos. 

## Import packages
1. pandas
2. numpy
3. matplotlib.pyplot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import the fpkms.csv file as a dataframe 
df_fpkms = pd.read_csv ("../../qbb2021/data/fpkms.csv")
```

## Extract and modify expression levels

```python
#get male_10 expression
male_10 = df_fpkms.loc[:, "male_10"]
#log transform male_10 expression
male_10_log2 = np.log2(male_10 + 1)
```

```python
#get male_14D expression
male_14D = df_fpkms.loc[:, "male_14D"]
#log transform male_14D expression
male_14D_log2 = np.log2(male_14D + 1)
```

## Calculate "A" and "M"

```python
A = 1/2 * (male_10_log2 + male_14D_log2) #see introduction for reference
M = male_10_log2 - male_14D_log2 #see introduction for reference
```

## Display the MA plot

```python
fig, ax = plt.subplots()
ax.scatter(A, M, s = 0.1)
ax.set_xlabel("Average gene abundance (A)")
ax.set_ylabel("Log ratios of gene expressions in two stages (M)")

fig.savefig("../../qbb2021-answers/day4-lunch/MA_plot.png")
```
